```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(8,5))

category_sales.plot(
    kind='bar',
    color=['red', 'green', 'blue']
)

plt.title('Category-wise Sales')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.xticks(rotation=0)

plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.month
                          )['Sales'].sum()

plt.figure(figsize=(10,5))

monthly_sales.plot(
    kind='line',
    marker='o',
    color='blue'
)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(10,5))

monthly_sales.plot(
    kind='line',
    marker='o',
    color='red'
)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.xticks(ticks=range(1,13), labels=months)

plt.grid(True)

plt.show()
```

## Monthly Sales Trend

### Objective
Analyze sales performance across different months.

### Key Findings

- Sales were lowest in February.
- Sales started increasing from March onwards.
- September, November, and December recorded the highest sales.
- November achieved the peak sales performance.

### Business Insights

- Sales exhibit clear seasonality.
- Year-end months generate significantly higher revenue.
- Customers tend to spend more during festive and holiday periods.

### Recommendations

- Increase inventory before peak months.
- Launch promotional campaigns before November and December.
- Allocate additional marketing budget during high-demand periods.
- Improve supply chain planning for seasonal demand spikes.

```python
import seaborn as sns
correlation = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()

plt.figure(figsize=(8,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

# Correlation Analysis

## Objective
To understand the relationship between Sales, Quantity, Discount, and Profit.

## Key Findings

- Sales and Profit have a moderate positive correlation (0.48).
- Discount and Profit have a negative correlation (-0.22).
- Quantity has a very weak relationship with Profit.
- Higher discounts may reduce overall profitability.

## Business Insight

The negative correlation between Discount and Profit suggests that excessive discounting can hurt business profits.

The company should optimize discount strategies to maintain profitability while attracting customers.

# Loss-Making Product Analysis

## Objective
To identify products that generate the highest losses for the business.

## Why This Matters

Not all products contribute positively to company profits.

Identifying loss-making products helps management:
- Review pricing strategies
- Reduce unnecessary discounts
- Improve inventory decisions
- Focus on profitable products

## Analysis
The following output shows the Top 10 products with the lowest total profit.

```python
df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
```

## Findings

The analysis identified several products generating significant losses.

### Key Observations

- Cubify CubeX 3D Printer Double Head Print recorded the highest loss (-$8,879.97).
- Multiple printer and conference table products appear among the top loss-making items.

### Business Impact
